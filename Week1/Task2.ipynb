```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings; warnings.simplefilter('ignore')
```

```python
train_df = pd.read_csv('titanic_train.csv')
test_df = pd.read_csv('titanic_test.csv')
```

```python
display(train_df)
```

```python
train_df.info()
```

```python
#splitting the Name to first and last name

train_df[['Name','Last']]=train_df['Name'].str.split(',', 1, expand=True)
train_df[['Honorofic','Last']] = train_df['Last'].str.split('.',1,expand=True)

test_df[['Name','Last']]=test_df['Name'].str.split(',', 1, expand=True)
test_df[['Honorofic','Last']] = test_df['Last'].str.split('.',1,expand=True)

display(train_df)
```

```python
#filling age voids with mean of that particular sex, embarked with most frequent, fare with mean grouped by sex
train_df['Age'] = train_df['Age'].fillna(train_df.groupby('Sex')['Age'].transform('mean'))
test_df['Age'] = test_df['Age'].fillna(test_df.groupby('Sex')['Age'].transform('mean'))

train_df['Embarked'] = train_df['Embarked'].fillna(train_df['Embarked'].value_counts().index[0])
test_df['Embarked'] = test_df['Embarked'].fillna(test_df['Embarked'].value_counts().index[0])

test_df['Fare'] = test_df['Fare'].fillna(test_df.groupby('Sex')['Fare'].transform('mean'))

train_df.info()
```

```python
#Cabin would be good information , but too many unknowns :(
train_df['CabinL'] = train_df['Cabin'].str[:1]
test_df['CabinL'] = test_df['Cabin'].str[:1]
train_df.info()
train_df=train_df.fillna('UNK')
test_df.info()
test_df=test_df.fillna('UNK')
```

```python
# for let in train_df['CabinL'].unique():
#     tf= train_df.loc[train_df['CabinL']==let]['Ticket']
#     print(let)
#     display(tf)
```

```python
Categorical_cols=[]
Numerical_cols =[]
for col in train_df.columns:
    if(train_df[col].dtype == 'O'):
        Categorical_cols.append(col)
    else:
        Numerical_cols.append(col)

Categorical_cols.remove('Ticket')
Categorical_cols.remove('Last')
Categorical_cols.remove('Cabin')

Numerical_cols.remove('PassengerId')
Numerical_cols.remove('Survived')

print(Numerical_cols)
print(Categorical_cols)
```

```python
X= train_df[Numerical_cols + Categorical_cols]
Y= train_df['Survived']
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder

X['CabinL'] = LabelEncoder().fit_transform(train_df['CabinL'])
X['Name'] = LabelEncoder().fit_transform(train_df['Name'])
X['Honorofic'] = LabelEncoder().fit_transform(train_df['Honorofic'])
X['Male'] = X['Sex'].replace({'male':1,'female':0})
X['Female'] = X['Sex'].replace({'female':1,'male':0})
X['Embarked'] = X['Embarked'].replace({'S':2,'C':1,'Q':0})


# column_to_encode = X['Honorofic']
# column_array = column_to_encode.values.reshape(-1, 1)
# encoded_array = OrdinalEncoder().fit_transform(column_array)
# X['Honorofic'] = encoded_array
X=X.drop('Sex', axis=1)
X=X.drop('Name',axis=1)
# X=X.drop('CabinL',axis=1)

display(X)
```

### DecisionTree, KNN, Naive Bayes, Ensemble Methods

```python
#DecisionTree

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#splitting of the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

dt_classifier = DecisionTreeClassifier(
    criterion='log_loss',  
    max_depth=None,  
    min_samples_split=2,  
    min_samples_leaf=2,  
    max_features=None,  
    random_state=42
)

# Fit the classifier to the training data
dt_classifier.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(
    n_estimators=120,  
    criterion='gini',  
    max_depth=None,  
    min_samples_split=2,  
    min_samples_leaf=2,  
    max_features=None, 
    random_state=42
)

# Fit the classifier to the training data
rf_classifier.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(
    n_neighbors=9, 
    weights='distance',  
    algorithm='auto',  
    p=1,
)

# Fit the classifier to the training data
knn_classifier.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = knn_classifier.predict(X_test)


accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
#Naive Bayes

from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()

# Fit the classifier to the training data
nb_classifier.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = nb_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
#Gradient Boost Classifier

from sklearn.ensemble import GradientBoostingClassifier

gb_classifier = GradientBoostingClassifier(n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    subsample=1.0,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='auto',
    random_state=42)

gb_classifier.fit(X_train, Y_train)

Y_pred = gb_classifier.predict(X_test)

gb_accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", gb_accuracy)
```

```python
#AdaBoost Classifier

from sklearn.ensemble import AdaBoostClassifier

adaboost_classifier = AdaBoostClassifier(n_estimators=700,
    learning_rate=1,
    algorithm='SAMME.R',
    random_state=42
)

adaboost_classifier.fit(X_train, Y_train)

Y_pred = adaboost_classifier.predict(X_test)

adaboost_accuracy = accuracy_score(Y_test, Y_pred)
print(" Accuracy:", adaboost_accuracy)
```
